```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("new_feature_table.csv")
```

```python
# Convert age to years (if not already present)
df["business_age_years"] = df["business_age_days"] / 365

# Create age groups
df["age_group"] = pd.cut(
    df["business_age_years"],
    bins=[0, 2, 5, df["business_age_years"].max()],
    labels=["New (0-2 yrs)", "Mid (2-5 yrs)", "Established (5+ yrs)"]
)
```

```python
# Plot boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x="age_group", y="stars", data=df)

plt.title("Ratings by Business Age Group")
plt.xlabel("Business Age Group")
plt.ylabel("Rating")

plt.tight_layout()
plt.show()
```

```python
df["age_group"] = pd.cut(
    df["business_age_years"],
    bins=[2, 5, df["business_age_years"].max()],
    labels=["Mid (2-5 yrs)", "Established (5+ yrs)"]
)
# Remove empty or unwanted group
df_filtered = df[df["age_group"] != "New (0-2 yrs)"]

# Plot again
plt.figure(figsize=(8,5))
sns.boxplot(x="age_group", y="stars", data=df_filtered)

plt.title("Ratings by Business Age Group")
plt.xlabel("Business Age Group")
plt.ylabel("Rating")

plt.tight_layout()
plt.show()
```
